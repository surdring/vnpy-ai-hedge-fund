# AI Agent Research Notebook

This notebook demonstrates how to use the vnpy AI Hedge Fund for research and backtesting.

```python
from vnpy_ai.data_adapter import DataAdapter
from vnpy_ai.workflow.runner import WorkflowRunner

# Initialize adapter and runner
adapter = DataAdapter()
runner = WorkflowRunner(adapter)

# Run analysis on a set of tickers
result = runner.run(tickers=["AAPL", "GOOGL", "MSFT"])
print(f"Decisions: {result.decisions}")
print(f"Degraded: {result.degraded}")
```
